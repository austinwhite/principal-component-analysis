```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
X = df.drop(["HeartDisease"], axis="columns")
y = df.HeartDisease
```

```python
encode = LabelEncoder()

encode.fit(X.Sex)
X.Sex = encode.transform(X.Sex)

encode.fit(X.RestingECG)
X.RestingECG = encode.transform(X.RestingECG)

encode.fit(X.ExerciseAngina)
X.ExerciseAngina = encode.transform(X.ExerciseAngina)

X.head()
```

```python
import numpy as np

np.unique(X.ST_Slope)
```

```python
np.unique(X.ChestPainType)
```

```python
X = pd.concat([X, pd.get_dummies(X.ChestPainType)], axis=1)
X = pd.concat([X, pd.get_dummies(X.ST_Slope)], axis=1)
X = X.drop(["ChestPainType", "ST_Slope"], axis="columns")
X.head()
```

```python
scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
vis = pd.DataFrame(X_scaled)
vis.head()
```

```python
X.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.96)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2)
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
